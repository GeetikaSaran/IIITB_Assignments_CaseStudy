__Defining a Function to plot Charts__

- As there are many columns to plot charts, defining a function with different plot/figure variables for further usage

```python
def plotting(x):        # here x is the column name
    plt.figure(figsize=(10,5))
    plt.subplot(1, 2, 1)
    sns.countplot(x, data=defaulters)
    plt.title('Distribution of '+ '%s' %x +' for Defaulters', fontsize=12)
    plt.xlabel(x)
    plt.xticks(rotation=90)
    plt.ylabel('Number of cases for defaulters')
    
    plt.subplot(1, 2, 2)
    sns.countplot(x, data=non_defaulters)
    plt.title('Distribution of '+ '%s' %x +' for Non-Defaulters', fontsize=12)
    plt.xlabel(x)
    plt.xticks(rotation=90)
    plt.ylabel('Number of cases for non-defaulters')
    
    plt.tight_layout()
    plt.show()
```

__Observations:__
- We can observe the maximum portion of data is not available for loan purpose
- We can also observe that there are more payment difficulties for the purpose of loan 'Repairs'
- Then bank/lender should have the correct data about purpose of loan, because a loan which is required for gambling or betting or some un-conventional investment practices may leads to higher losses to client and will create payment difficulties
